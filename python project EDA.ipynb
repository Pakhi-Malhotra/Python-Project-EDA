# Exploratory Data Analysis


# Gender

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#On the basis of Amount
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
chart=sns.barplot(data=sales_gen,x='Gender',y='Amount')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Amount v/s Gender',fontdict={'size':25,'weight':'bold'})
```

*From above graphs we can see that most of the buyers are females and even the amount spent by females is more than males.*

# Age Group

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#On the basis of Amount
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
chart=sns.barplot(data=sales_age,x='Age Group',y='Amount')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Amount v/s Age Group',fontdict={'size':25,'weight':'bold'})
```

*From above graphs we can see that most of the buyers are of age group between 26-35 years and the maximum amount spent on shopping is also by the people of age group between 26-35 years.*

# State

```python
df.columns
```

```python
#total no. of orders from top 10 states
sns.set(rc={'figure.figsize':(15,5)})
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
chart=sns.barplot(data=sales_state,x='State',y='Orders')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Orders v/s State',fontdict={'size':25,'weight':'bold'})
```

```python
#total amount/sales from top 10 states
sns.set(rc={'figure.figsize':(15,5)})
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
chart=sns.barplot(data=sales_state,x='State',y='Amount')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Amount v/s State',fontdict={'size':25,'weight':'bold'})
```

*From above graphs we can see that the most orders and the maximum amount spent on orders are from Uttar Pradesh, Maharashtra and Karnataka respectively.*

# Marital Status

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

*From this graph we can see that married people have done more shopping than unmarried people.*

```python
# On the basis of Amount
sales_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
chart=sns.barplot(data=sales_ms,x='Marital_Status',y='Amount',hue='Gender')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Amount v/s Marital Status',fontdict={'size':25,'weight':'bold'})
```

*Hence from the above graphs we conclude that most of the buyers among married people are women and that they spend more amount than men.*

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# On the basis of Amount
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
chart=sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Amount v/s Occupation',fontdict={'size':25,'weight':'bold'})
```

*From the above graphs we can see that most of the buyers are working in IT,Healthcare and Aviation Sector.*

# Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#On the basis of Amount
sns.set(rc={'figure.figsize':(20,5)})
sales_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
chart=sns.barplot(data=sales_pc,x='Product_Category',y='Amount')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Amount v/s Product Category',fontdict={'size':25,'weight':'bold'})
```

*From the above graphs we can see that the most orders were from Clothing and Apparel,Food and Electronics and Gadgets respectively but the maximum amount spent is on Food,Clothing and Apparel and Electronics and Gadgets respectively.*

# Product ID

```python
df.columns
```

```python
# On the basis of Orders
sns.set(rc={'figure.figsize':(20,5)})
sales_pid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
chart=sns.barplot(data=sales_pid,x='Product_ID',y='Orders')
for bars in chart.containers:
    chart.bar_label(bars)
chart.set_title('Orders v/s Product ID',fontdict={'size':25,'weight':'bold'})
```
